```python
def is_prime(n):
    if n<=1:
        return False
    for i in range(2,n):
        if n%i==0:
            return False    
    return True


num = 11
if is_prime(num):
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
print(is_prime(3))
print(is_prime(7))

```

```python
def is_prime1(n):
    for i in  range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
print(is_prime1(22))
print(is_prime1(12))
print(is_prime1(2))
```

```python
def get_square(n):
    return n**2
print(get_square(10))

fun=lambda n:n**2
print(fun(20))

print((lambda n:n**2)(30))

print((lambda a,b: a+b)(100,200))

print((lambda n: "Even" if n%2==0 else "odd" )(30))
```

#### * args *kwargs

```python
def get_sum(*args):
    print(args,type(args))
    result=0
    for i in args:
        result+=i
    return result

print(get_sum(100))
print(get_sum(90,30,20))
```

```python
def get_total_budget(**kwargs):
    print(kwargs,type(kwargs))
    total=sum(kwargs.values())
    print("Sum is :",total)


get_total_budget(amol=1000)
get_total_budget(amol=111,mil=345,gilk=876)
```
