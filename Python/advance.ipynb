# Exception Handling and traceback

```python
# there are four block in exception handling
# try,except,else,finally
n=int(input("Enter the number"))
result=100/n
print(f"Result is:{result} :")
```

```python
try:
    n=int(input("Enter the number"))
    result=100/n
    print(f"Result is :{result}")
    n()
except ZeroDivisionError as e:   
    print(f"You cannot divide by zero:{e}")
except ValueError as e:
    print(f"Invalid input , please enter a valid number {e}")
except Exception as e:
    print(f"General exception {e}")
else:
    print("Inside else block")
finally:
    print("Inside finally block ")

```

```python
import traceback
try:
    n=int(input("Enter the number"))
    result=100/n
    print(f"Result is :{result}")
except ZeroDivisionError as e:   
    print(f"You cannot divide by zero:{e}")
    traceback.print_exc()
except ValueError as e:
    print(f"Invalid input , please enter a valid number {e}")
    traceback.print_exc()
except Exception as e:
    print(f"General exception {e}")
    traceback.print_exc()
else:
    print("Inside else block")
finally:
    print("Inside finally block ")

```

```python
import traceback 

class AppError(Exception):
    """base class for application exception""" # it is called doc string which is inside"""""""
    pass

class ValidationError(AppError):
    def _init_(self,message):
        super()._init_(message)

try:
    n=input("Enter number :")
    if not n.isnumeric():
        raise ValidationError("Invalide numeric input from the user.")
    print("you entered the valide number: ",n)
    
except ValidationError as e:
    print(e)
    traceback.print_exc()

```

```python
def get_value():
    yield 10  # it become generator when we add yield keyword
    yield 20
    yield 30

print(type(get_value))
print(type(get_value()))
g=get_value()
print(next(g))
print(next(g))  
print(next(g))  
```

```python
def my_range(start,end,step):
    while start<end:
        yield start
        start+=step


for n in my_range(1,10,0.5):
    print(n,end="")
```

```python
from random import random
class Myrandom:
    def __init__(self,low,high,count):
        self.low=low
        self.high=high
        self.count=count
        self.counter=0 # To keep a track of number of generated values

    def __next__(self):
        if self.counter<self.count:
            self.counter+=1
            return (self.high-self.low)*random()+self.low
        else:
            raise StopIteration
        
    
    def __iter__(self):
        return self
    


for n in Myrandom(10,25,10):
    print(n)
    
    

    
```

#### Decorators


```python
def dec(fun):
    print("Hello World !")
    fun()
    print("Welcome to Python ")

def greet():
    print("Hello World in python ")

def welcome():
    print("Welcome to Python world")

dec(welcome)
dec(greet)
```

```python
def dec(fun):
    def wrapper():
        print(f"Before calling {fun.__name__}")
        fun()
        print(f"After calling  {fun.__name__}")
    return wrapper


@dec
def greet():
    print("Hello World in python ")

@dec
def welcome():
    print("Welcome to Python world")

welcome()
greet()
```

```python
def dec(fun):
    def wrapper(*args,**kwargs):
        print(f"Before calling {fun.__name__}")
        fun(*args,**kwargs)
        print(f"After calling  {fun.__name__}")
    return wrapper


@dec
def greet(name):
    print(f"Hello {name}")
greet('abc')

@dec
def welcome():
    print("Welcome to Python world")

@dec 
def sum(a,b):
    print(f"{a}+{b}= ",a+b)

welcome()
sum(20,10)
```

```python
def dec(fun):
    t1=time.time()
    def wrapper():
        print(f"Before calling {fun.__name__}")
        fun()
        print(f"After calling  {fun.__name__}")
    return wrapper


@dec
def greet():
    print("Hello World in python ")
    t2=time.time()
    print(t2-t1)

greet()
```

```python
def time_it(fun):
    def wrapper():
        t1=time.time()
        print(f"Before calling {fun.__name__}")
        fun()
        t2=time.time()
        print(f"After calling  {fun.__name__}")
        print(f"Time taken to execute {fun.__name__} is {t2-t1:0.6f} seconds")
    return wrapper

@time_it
def greet():
    print("Hello World in python ") 
    for i in range(1000000):
        pass

greet()
```

```python
def timecal():
    t1=time.time()
    print(f"before time is: {t1}")
    for i in range (10):
        print(i)
    t2=time.time()
    print(f"After time is: {t2}")
    print(f"Differnce between them :{t2-t1}")

timecal()
```

```python
def repeat(n):
    def decorator(fun):
        def wrapper(*args ,**kwargs):
            print(f"before calling {fun.__name__} function")
            for i in range(n):
                fun(*args,**kwargs)
            print(f"After Calling {fun.__name__} function ")
        return wrapper
    return decorator

@repeat(3)
def greet(name):
    print(f"Hello world! {name}")

greet('ovi')
```

 #### Contex Manager  
 ##### context manager used to close file

```python

# __function() dunder methodes get call automaticaly

class MyFileHandler:
    def __init__(self,filename,mode):
        self.filename=filename
        self.mode=mode 

    def __enter__(self):
        print("Entering the contex.....")
        self.file=open(self.filename,self.mode)
        return self.file

    def __exit__(self,exc_type,exc_val,exc_tb):
        if self.file:
            self.file.close()
        print("Exiting from the contex :")

with MyFileHandler("abc.txt",'w') as f:
    f.write("Hello world !")
    
```
